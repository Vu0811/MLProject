```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Get USA Housing data
train = pd.read_csv(r'E:/Works/CODE/Artificial Intelligence/Data/Linear-Regression/USA_Housing.csv')
train = train[:3000]
train_x = np.array([train['Avg. Area Income'],train['Avg. Area House Age'],train['Avg. Area Number of Rooms'],train['Avg. Area Number of Bedrooms'],train['Area Population']]).T
train_y = np.array([train['Price']]).T
```

```python
#Get train data
train = pd.read_csv (r'E:/Works/Visual Studio/Artificial Intelligence/Data/Linear-Regression/train.csv')
train_x = np.array([train['x']]).T
train_y = np.array([train['y']]).T
```

```python
#Display train data
plt.plot(train_x, train_y, 'ro')
plt.axis([0, 100, 0, 100])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Linear-Regression Algorithm
# Building Xbar 
one = np.ones((train_x.shape[0], 1))
Xbar = np.concatenate((one,train_x), axis = 1)
#Get w = A.b
A = np.dot(Xbar.T, Xbar)
b = np.dot(np.transpose(Xbar), train_y)
w = np.dot(np.linalg.pinv(A), b)
```

```python
# Preparing the fitting line 
w_0 = w[0][0]
w_1 = w[1][0]
x0 = np.linspace(5, 95, 2)
y0 = w_0 + w_1*x0
```

```python
# Drawing the fitting line 
plt.plot(train_x.T, train_y.T, 'ro')     # data 
plt.plot(x0, y0)               # the fitting line
plt.axis([0, 100, 0, 100])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
